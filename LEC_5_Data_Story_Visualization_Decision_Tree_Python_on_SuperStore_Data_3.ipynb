**Descriptive Statistics:**

1. Count: Count represents the number of observations or data points in a dataset. It indicates the total number of values in the dataset, excluding missing or NaN (Not-a-Number) values.

2. Mean (Average): The mean is the arithmetic average of all the values in the dataset. It's calculated by adding up all the values and then dividing by the count. The mean provides a measure of the dataset's central tendency.

3. Standard Deviation (std): The standard deviation measures the degree of variation or dispersion in a dataset. It quantifies how spread out the values are from the mean. A higher standard deviation indicates greater variability.

Easier guide to STANDARD DEVIATION:
https://www.statology.org/what-is-a-good-standard-deviation/

To clarify:

FOR EXAMPLE: our unit is year, std is 8.

If the mean is 2014 and the standard deviation is 8, you would expect that  assuming a roughly normal distribution  about 68% of the data points fall between 2006 (2014 - 8) and 2022 (2014 + 8).

This does not mean all data points are within this range  some may be much closer or farther. The standard deviation measures the typical spread around the mean, with larger standard deviations indicating greater dispersion.

The standard deviation quantifies the spread or dispersion of your data. A smaller standard deviation indicates that data points tend to be closer to the mean, while a larger standard deviation indicates that data points are more spread out from the mean.

So, in summary, a standard deviation of 8 suggests that data points are, on average, 8 units above or below the mean, but individual data points may vary from this average.

4. Minimum (min): The minimum is the smallest value in the dataset. It represents the lowest data point in the distribution.

5. 25% (First Quartile): This value represents the 25th percentile of the data. It's the value below which 25% of the data falls. It's also known as the first quartile (Q1).

6. 50% (Median or Second Quartile): The 50th percentile, also known as the median or second quartile (Q2), is the middle value when the data is sorted in ascending order. It divides the dataset into two equal halves.

7. 75% (Third Quartile): This value represents the 75th percentile of the data. It's the value below which 75% of the data falls. It's also known as the third quartile (Q3).

8. Maximum (max): The maximum is the largest value in the dataset. It represents the highest data point in the distribution.

**Cross-Tabulation:**

Cross-tabulation, often referred to as a contingency table or crosstab, is a statistical technique used to summarize and analyze the relationship between **two categorical variables.** It provides a way to examine how the values of one variable are distributed across the categories of another variable. Here's a brief explanation:

***Variables:*** In a crosstab, you have two categorical variables, typically referred to as the row variable and the column variable.

***Frequency Counts:*** The main purpose of cross-tabulation is to calculate and display the frequency counts of how many observations fall into each combination of categories for the two variables. Each cell in the table represents a unique combination of category values.

***Summarizing Relationships:*** By creating a cross-tabulation, you can quickly see how the distribution of one variable varies or depends on the categories of the other variable. It helps identify patterns, associations, or dependencies between the two variables.

*Example:* For instance, if you have data on the gender and buying preferences of customers, a cross-tabulation could show you how many male and female customers prefer different product categories. This can help in market segmentation and targeted marketing.

```
def find_outliers_IQR(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    outliers = series[(series < lower) | (series > upper)]
    return outliers
```

You can do further research on when you can remove your outliers. But here is my guide:

**Reasons to Remove Outliers:**

  Data Quality: Outliers may result from errors in data collection or recording, and they can represent data quality issues. Removing them can help improve the overall quality of your dataset.

  Statistical Assumptions: Many statistical methods assume that data is normally distributed or follows certain patterns. Outliers can violate these assumptions, leading to biased or inaccurate results.

  Visualizations: Outliers can distort data visualizations, making it challenging to interpret or communicate results effectively. Removing them can lead to clearer and more informative visualizations.


**When to Remove Outliers:**

  Domain Knowledge: Consider the context and domain knowledge. Some outliers may be valid and meaningful data points, especially in certain research or industry-specific contexts. Be cautious when removing outliers and ensure it aligns with the problem you're trying to solve.

  Data Goals: The decision to remove outliers should be based on the goals of your analysis. If your goal is to understand the overall trend and characteristics of the data, removing outliers might be appropriate. However, if you want to specifically analyze extreme values, keeping them may be more relevant.


**How to Remove Outliers:**

  Deletion: Simply remove the outlier data points from the dataset. This is the most straightforward method but can result in loss of information.

  Transformation: Apply mathematical transformations to the data (e.g., logarithmic transformation) to reduce the impact of outliers without completely removing them.

  Imputation: Replace outliers with more typical values, often using techniques like mean imputation or interpolation.

You can also set a range to remove outliers: this case I filtered out the dataframe to only keep the rows with income less than 140000
