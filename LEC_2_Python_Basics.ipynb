## LIST METHODS

```python
my_list = [1, 2, 3, 4, 5]
```

```python
# append(item): Adds an item to the end of the list

my_list.append(6)
my_list
```

```python
# extend(item): Appends the elements of an iterable to the end of the list

my_list.extend([7, 8])
my_list
```

```python
# insert(index, item): Inserts an item at a specific position in the list

my_list.insert(3, 9)
my_list
```

```python
# remove(item): Removes the first occurrence of the specified item from the list

my_list.remove(2)
my_list
```

```python
# sort(): Sorts the list in ascending order

my_list.sort()
my_list

#use reverse = True for descending order inside ()
```

```python
# reverse(): Reverses the order of elements in the list

my_list.reverse()
my_list
```

```python
# pop(index): Removes and returns the item at the specified index

my_list.pop(0)
my_list
```

```python
# clear(): Removes all items from the list, making it empty

my_list.clear()
my_list
```

```python
my_list = [1, 2, 3, 4, 5]
```

```python
# index(item): Returns the index of the first occurrence of the specified item

index = my_list.index(5)
index
```

```python
# count(item): Returns the number of occurrences of the specified item

count = my_list.count(4)
count
```

```python
# copy(): Returns a shallow copy of the list

copy_of_list = my_list.copy()
copy_of_list
```

```python
# len(list): Returns the number of items in the list

length = len(my_list)
length
```

```python
# max(list): Returns the maximum item in the list

maximum = max(my_list)
maximum
```

```python
# min(list): Returns the minimum item in the list

minimum = min(my_list)
minimum
```

```python
# sum(list): Returns the sum of all items in the list (if they are numeric)

total = sum(my_list)
total
```

##Creating and Merging Dataframes:

**Inner Join:**

→ An inner join returns only the rows that have matching values in both DataFrames based on the specified key(s).

→ In the merged DataFrame, only the rows where the key(s) from both DataFrames match are included.

**Right Join:**

→ A right join returns all the rows from the right DataFrame and the matched rows from the left DataFrame.

→ If there are no matching rows from the left DataFrame, NaN values are filled in for the columns from the left DataFrame.

**Left Join:**

→ A left join returns all the rows from the left DataFrame and the matched rows from the right DataFrame.

→ If there are no matching rows from the right DataFrame, NaN values are filled in for the columns from the right DataFrame.

**Outer Join:**

→ An outer join returns all rows from both DataFrames, with NaN values where there is no match.

→ This includes all rows from both DataFrames, even if there are no matching rows in the other DataFrame.

image.png

```python
import pandas as pd

# create product dataframe

product = pd.DataFrame({
  "product_code": [1001, 1002, 1003, 1004],
  "weight": [125, 200, 100, 400],
  "price": [10.5, 24.5, 9.9, 34.5]
})
product
```

```python
# create product dataframe

sales = pd.DataFrame({
  "product_code": [1001, 1002, 1003, 1007],
  "sales_date": ["2021-12-10"] * 4,
  "sales_qty": [8, 14, 22, 7]
})
sales
```

```python
# merge DataFrames
merged_df_left = product.merge(sales, how="left", on="product_code")

merged_df_left
```

```python
# merge DataFrames
merged_df_right = product.merge(sales, how="right", on="product_code")

merged_df_right
```

```python
# merge DataFrames
merged_df_inner = product.merge(sales, how="inner", on="product_code")

merged_df_inner
```

```python
# merge DataFrames
merged_df_outer = product.merge(sales, how="outer", on="product_code")

merged_df_outer
```
