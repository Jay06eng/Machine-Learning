#Mean, Median, Standard Deviation, and Correlation Coefficient

```python
import numpy as np

# The dataset
learning_hours = [1, 2, 6, 4, 10]
scores = [3, 4, 6, 5, 6]
```

##Mean and Median

Both are measures of central tendency

*   Mean - The mean, often referred to as the average, is calculated by adding up all the values in a data set and then dividing the sum by the total number of values.
*   Median - The median is the middle value of a data set when it is ordered from least to greatest. If there is an even number of observations, the median is the average of the two middle values.



```python
# Applying some stats methods to understand the data:
print("Mean learning time: ", np.mean(learning_hours))
print("Mean score: ", np.mean(scores))
print("Median learning time: ", np.median(learning_hours))
print("Median scores: ", np.median(scores))
```

## Standard Deviation


*   how much the data is spread out around the mean

--- “Are all the scores close to the average?” <br>
--- "Are lots of scores way above or way below the average score?"

```python
print("Mean learning time: ", np.mean(learning_hours))
print("Standard deviation: ", np.std(learning_hours))
```

```python
print("Mean score: ", np.mean(scores))
print("Standard deviation: ", np.std(scores))
```

##**VERBAL INTERPRETATION:**<br><br>


**-Learning Hours:**

The average learning time is 4.6 hours. Combining this with the standard deviation of 3.2, it means that most students study around 4.6 hours, and the typical difference from this average is about 3.2 hours. So, some students might study a bit more or less, but the majority are within around 3.2 hours of the average study time.

**-Scores:**

The average score is 4.8. Combined with the standard deviation of 1.17, it means that most students score around 4.8, and the typical difference from this average is about 1.17 points. So, some students might score a bit higher or lower than 4.8, but the majority are within around 1.17 points of the average score.


*In summary, the means tell you where the center of your data is, and the standard deviations give you an idea of how spread out or clustered the individual data points are around these averages.*

##Question 2:

Is correlation causation? Why? Why not?

For example, say an ice-cream shop has data for how many sunglasses were sold by a big store in the neighborhood over a period of time, and they decide to compare sunglasses sales to their ice cream sales. From their results, they find a high correlation between the sales of the two. Does this mean that sunglasses make people want to buy ice cream? Hmm, no!

So, in layman terms, *“Correlation Is Not Causation”* tries to remind us that correlation does not prove one thing/event causes the other, but:

• One event might cause the other <br>
• The other might cause the first to happen<br>
• They may be linked by a different reason<br>
• Or the result could have been random chance<br>
