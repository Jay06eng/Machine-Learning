```python
# I create a new column by telling Python to use 2015 and substract the values
# in the dataframe's column "Year_Birth". Then display to see the first 5 rows
# of the new dataframe (same dataframe plus the new Age named "Age")

df['Age'] = 2015 - df['Year_Birth']
display(df.head())
```

```python
df['Age'] = 2015 - df['Year_Birth']

df

```

```python
#Now that I have the age column, I can classify the age into buckets or categories using the "bin sizes" given in the exercise instructions.

#'Gen Z (0-24)': Ages from 0 to 24 (inclusive)
#'Gen Y (25-39)': Ages from 25 to 39 (inclusive)
#'Gen X (40-54)': Ages from 40 to 54 (inclusive)
#'Older (55-59)': Ages from 55 to 59 (inclusive)
#'Senior (60+)': Ages 60 and above

#what is the rationale of pd.cut()
# pandas function used to segment and sort data values into bins. It's
# particularly useful for converting continuous numerical data into discrete
# categorical bins. We used it here, to take continuous data from Age column and
# "cut" it into the bins defined in the exercise instructions.
#The purpose of 'inf' is to create an open-ended bin for "Senior(60+)" category
#Since Senior cat includes anything beyond 60 and no upper limits, float('inf') ensures
#any age greater than 60 either a float or not goes into that bin.


bins = [0, 24, 39, 54, 59, float('inf')]
labels = ['Gen Z (0-24)', 'Gen Y (25-39)', 'Gen X (40-54)', 'Older (55-59)', 'Senior (60+)']
df['AgeCat'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
display(df.head())

df
```

```python
bins = [0, 24, 39, 54, 59, float('inf')]
labels = ['Gen Z (0-24)', 'Gen Y (25-39)', 'Gen X (40-54)', 'Older (55-59)', 'Senior (60+)']
new_var = False
df['AgeCat'] = pd.cut(df['Age'], bins=bins, labels=labels, right=new_var)
display(df.head())

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='AgeCat')
plt.title('Distribution of Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#Nov 10 class review
#sorted by age
#another chart would be a cluster chart (Like below one
#but with an extra column) using the Agecat
# and the total spending/income

#Another chart indicating the most bought
#items, e.g. fruits are best sellers, bundle
#those with lower amount of bought products
#to increase revenue by enticing customers
#to buy those less bought products.
```

```python
#I create a new column for total spending
#The new column will give me the ability to compare total spending with other columns like agecat, income, education.

spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalSpending'] = df[spending_cols].sum(axis=1)

spending_by_agecat = df.groupby('AgeCat', observed=False)['TotalSpending'].sum().reset_index()
display(spending_by_agecat)
```

```python
#the above works but is given me the total spending by AgeCat
# Add a new column called 'TotalSpending' that sums spending across all product categories
df['TotalSpending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)

# Display first 5 rows to confirm
print(df[['Id', 'Age', 'TotalSpending']].head())
```

```python
#Discovering Age distribution

plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Created a distribution using the Age Category
#for a better understanding of the age data.


plt.figure(figsize=(10,6))
sns.histplot(df['AgeCat'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('AgeCat')
plt.ylabel('Frequency')
plt.show()
```

```python
#Comparison between Education versus Income

grouped_df = df.groupby('Education')['Income'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Education', y='Income', data=grouped_df)
```

```python

pivot_agecat_spending = df.pivot_table(index='AgeCat', values='TotalSpending', aggfunc='sum')
pivot_agecat_spending
```

```python
#histogram of total spending
plt.figure(figsize=(10,6))
sns.histplot(df['TotalSpending'], bins=20, kde=True)
plt.title('Histogram of Total Spending per Age Category')
plt.xlabel('Total Spending')
plt.ylabel('Customers')
plt.show()
```

```python
#Bar chart Total Spending by Agecat from pivot
plt.figure(figsize=(8,5))
pivot_agecat_spending.plot(kind='bar')
plt.title('Total Spending by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Total Spending')
plt.show()
```

```python
#Line plot
age_avg_spending = df.groupby('AgeCat')['TotalSpending'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(x='AgeCat', y='TotalSpending', data=age_avg_spending)
```

```python
#Scatter Plot Income versus Total Spending
plt.figure(figsize=(8, 5))
plt.scatter(df['Income'], df['TotalSpending'], alpha=0.4)
plt.title('Income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.tight_layout()
plt.show()
```
