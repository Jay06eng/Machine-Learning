```python
#Converting 'album_release_date' to datetime objects with error handling
df['album_release_date'] = pd.to_datetime(df['album_release_date'], errors='coerce')

# Check for NaT values (unparseable dates)
nat_count = df['album_release_date'].isna().sum()
print(f"Number of unparseable dates (NaT): {nat_count}")

# Extract year, month, day, week, and quarter from the corrected datetime column
df['album_release_year'] = df['album_release_date'].dt.year
df['album_release_month'] = df['album_release_date'].dt.month
df['album_release_day'] = df['album_release_date'].dt.day
df['album_release_day_name'] = df['album_release_date'].dt.isocalendar().week
df['album_release_quarter'] = df['album_release_date'].dt.quarter

# Drop the original album_release_date column as its components are now extracted
df = df.drop('album_release_date', axis=1)

# Display the head of the DataFrame to see the new columns
display(df.head())

# Display unique years to verify the parsing
print("\nUnique album release years after correction:")
print(df['album_release_year'].dropna().astype(int).unique())
```

```python
# 1952-1972 = Oldies
# 1973-1992 = Classics
# 1993-2012 = Recent
# 2013-2025 = New_ones

#Convert album release column to numeric
df['album_release_year'] = pd.to_numeric(df['album_release_year'], errors='coerce')

df['track_age_cat'] = pd.cut(df['album_release_year'],
                                     bins=[1952, 1972, 1992, 2012, 2025],
                                     labels=['Oldies', 'Classics', 'Recent', 'New_ones'],
                             include_lowest=True)
df.head()

```

```python
df.dtypes
```

```python
df['track_duration_min'].describe()
```

```python
# 0-3.00 = too short
# 3.01-4.00 = ok
# 4.01+ = too long

df['track_duration_cat'] = pd.cut(df['track_duration_min'],
                                     bins=[0, 3.00, 4.00, float('inf')],
                                     labels=['too short', 'ok', 'too long'],
                             include_lowest=True)
df.head()
```

```python
# 0-30 = no body
# 31-60 = local
# 61-85 = famous
# 86-100 = ultra_famous
df['artist_popularity_cat'] = pd.cut(df['artist_popularity'],
                                     bins=[0, 30, 60, 85, 100],
                                     labels=['no body', 'local', 'famous', 'ultra_famous'],
                             include_lowest=True)
df.head()
```
