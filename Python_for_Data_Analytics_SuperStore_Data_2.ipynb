#The Cut and Qcut Functions

**CUT**
<br>
The cut function divides the entire value range into intervals of the same size, called bins. The range covered by each bin will be the same.

````python

# Create a Series with values between 0 and 25
A = pd.Series([5, 0, 2, 8, 4, 10, 7 , 25, 19, 13])
A
````

````python
# cut function
A_binned = pd.cut(A, bins=5)

A_binned
````

````python
 A_binned.cat.categories
````

In the interval notation used by Pandas, the round bracket "(" indicates that the value is not included in the interval, while the square bracket "[" indicates that the value is included.

So, in the interval (-0.025, 5.0], the value -0.025 is not included in the interval, but the value 5.0 is included.

###What pd.cut() does

pd.cut() divides continuous numeric data into equal-width bins.

You used bins = 5, so pandas:

1. Looks at the minimum value: 0
2. Looks at the maximum value: 25
3. Splits the full range (0 → 25) into 5 equal-width intervals.

Range = 25 − 0 = 25

Bin width = 25 / 5 = 5

So your 5 bins are:

- (–0.025, 5]
- (5, 10]
- (10, 15]
- (15, 20]
- (20, 25]

*Note: The slight negative value is a safety offset so the minimum value is included.*

let's review the output:

````python
# cut function
A_binned = pd.cut(A, bins=5)

A_binned
````

Customize your bins if:

- You want full control over bin boundaries
- You don’t want pandas to guess the ranges
- You want to include specific breakpoints
- You want meaningful bins (e.g., score brackets, risk levels)

````python

# Create a Series with values between 0 and 25
A = pd.Series([5, 0, 2, 8, 4, 10, 7 , 25, 19, 13])

# cut function
A_binned = pd.cut(A, bins=[-1, 5, 10, 15, 20, 25])

print(A_binned)
````

````python

# Create a Series with values between 0 and 25
A = pd.Series([5, 0, 2, 8, 4, 10, 7 , 25, 19, 13])

# cut function
A_binned = pd.cut(A, bins=[-1, 5, 10, 15, 20, 25],
                  labels=['0-5', '6-10', '11-15', '16-20', '21-25'])

print(A_binned)
````

##Qcut
<br>
The qcut function groups the values into bins so that there is approximately the same number of values in each bin. The upper and lower bounds are adjusted accordingly. The following block of code demonstrates how the qcut function works.

````python
# The qcut function
A = pd.Series([1, 4, 2, 10, 5, 6, 8, 7, 5, 3, 5, 9])

A_binned = pd.qcut(A, q=3)

print(A_binned.value_counts())
````

pd.qcut() divides the data into quantile-based bins, meaning:

- Each bin contains (approximately) the same number of observations
- Bins are created based on the distribution of the data, not equal-sized ranges

<br>

**cut** splits by number ranges

**qcut** splits by number of rows

#The "where" Function

 The where function evaluates the given condition and then updates the values. When the condition is True, the original values are kept the same. When it’s False, the values are updated using the other parameter.

`DataFrame.where(condition, other=default_value) `

#LINE PLOT

The x-axis of a line plot usually represents time with varying frequencies, such as an hour, day, month, year, and so on. The y-axis shows a quantity, so the **line plot essentially tracks a quantity’s change over time.**

When the data is timeseries, this method is good.

First, I will create a made up data since a time series data is not applicable in our dataset.

````python
import numpy as np
````

````python
# Creating a made-up time series with a range of dates
start_date = '2000-01-01'
end_date = '2000-12-31'
date_range = pd.date_range(
    start=start_date, end=end_date, freq='M')  # Monthly frequency

# Creating a DataFrame with the date range and a corresponding made-up 'MntWines' column
df = pd.DataFrame({
    'Date': date_range,
    'MntWines': np.random.randint(500, 1500, size=len(date_range))  # Random values for 'MntWines'
})
````

500 → the lowest value (inclusive, meaning it can be 500).

1500 → the highest value (exclusive, meaning it will go up to 1499, but not 1500).

size=len(date_range) → how many random numbers to generate (one for each date in date_range).

````python
df.plot(
    x = "Date",
    y = "MntWines",
    kind = "line",
    figsize = (10,6),
    title = "Amount spent on wine over time",
    xlabel = "Date",
    ylabel = "Amount spent on wine"
    )
````

````python
avg_wine = df.groupby('Year_Birth')['MntWines'].mean()
avg_wine
````

````python
avg_wine.plot(
    x = "Year_Birth",
    y = "MntWines",
    kind = "line",
    figsize = (10,6),
    title = "Amount spent on wine per year of birth",
    xlabel = "Year of Birth",
    ylabel = "Amount"

)
````

````python
grouped_df = df.groupby(
    'Year_Birth')[['MntWines', 'MntFruits']].mean()
grouped_df
````

````python
grouped_df.plot(
    x = "Year_Birth",
    y = "MntWines",
    kind = "line",
    figsize = (10,6),
    title = "Amount spent on wine per year birth",
    xlabel = "Year of Birth",
    ylabel = "Amount spent on wine",
    secondary_y = "MntFruits"
    )
````

When you use groupby and then plot on the resulting DataFrame, you don't need to specify the x and y parameters directly in the plot call.
<br> In the following code:

grouped_df.plot() is called directly on the DataFrame resulting from the groupby operation. The x-axis will automatically be set to the index of the DataFrame (in this case, 'Year_Birth'). The secondary_y parameter is used to specify the secondary y-axis for the 'MntFruits' column.

````python
grouped_df.plot(
    kind='line',
    figsize=(10, 6),
    title="Amount spent on wine and fruits per year of birth",
    xlabel="Year of Birth",
    ylabel="Amount spent",
    secondary_y="MntFruits")
````

````python
grouped_df.plot(
    kind='line',
    figsize=(10, 6),
    title="Amount spent on wine and fruits per year of birth",
    xlabel="Year of Birth",
    ylabel="Amount spent")
````
