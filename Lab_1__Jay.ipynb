````python
import pandas as pd

data1 = {'class_id': [6, 7, 6, 6, 8],
         'class_name': ['Math', 'Science', 'Math', 'Math', 'PE']}
df1 = pd.DataFrame(data1)

data2 = {'id': [1, 2, 3, 4, 5],
         'name': ['student1', 'student2', 'student3', 'student4', 'student5'],
         'class_id': [6, 7, 6, 6, 8]}
df2 = pd.DataFrame(data2)

# Merge them to get your main dataframe
df = df2.merge(df1, how="left", on="class_id")
display(df)


````

````python
df['name'] = df['name'].astype(str)
df['name'] = df['name'].str.replace('student1', 'Anthony')
df['name'] = df['name'].str.replace('student2', 'Gail')
df['name'] = df['name'].str.replace('student3', 'Nick')
df['name'] = df['name'].str.replace('student4', 'Nomar')
df['name'] = df['name'].str.replace('student5', 'Teresa')

display(df)
````

````python
# Recreate df by merging df1 and df2
df = df2.merge(df1, how="left", on="class_id")
display(df)
````
