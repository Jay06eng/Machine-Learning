#Part 2: Correlation Coefficients

Interpret the following in terms of mean, std, and correlation coefficient:

```python
import numpy as np

# Manually create a dataset with 15 observations
coffee_consumption = [2, 4, 1, 5, 3, 1, 3, 2, 4, 5, 1, 4, 3, 2, 5]
productivity_scores = [60, 75, 50, 80, 70, 55, 65, 60, 72, 85, 48, 78, 68, 58, 90]
```

```python
print("the mean for  coffee consumption is ", np.mean(coffee_consumption))
print("the mean for  coffee consumption is ", np.mean(productivity_scores))
print("the standard deviation for coffee consumption is " , np.std(coffee_consumption))
print("the standard deviation for productivity scores is " , np.std(productivity_scores))
np.corrcoef(coffee_consumption, productivity_scores)
```

The correlation is 0.97, which is very close to 1.

This means as coffee consumption goes up, productivity almost always goes up too in this dataset but it doesnt prove coffee causes.
