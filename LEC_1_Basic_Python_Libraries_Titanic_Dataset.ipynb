```python
from IPython.core.display import publish_display_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('TITANIC DATASET.csv')

df
```

```python
#The exercise is asking to create variable(s) for a
#categorical data set in a column for a binary value
#in preparation for our statistical analysis.
#Pclass = [1:1st, 2:2nd, 3:3rd]
```

```python
# Map the numerical 'Pclass' values to their string representations
pclass_mapping = {1: '1st', 2: '2nd', 3: '3rd'}
df['Pclass_category'] = df['Pclass'].map(pclass_mapping)

# Display the first few rows of the DataFrame with the new 'Pclass_category' column
display(df[['Pclass', 'Pclass_category']].head())
```

```python
df
```

```python
embarked_mapping = {'C':'Cherbourg', 'Q':'Queenstone', 'S':'Southampton'}
df['Embarked'] = df['Embarked'].map(embarked_mapping )
df
```

```python
# Check for missing values in each column and sum them up
missing_values = df.isnull().sum()

# Display the count of missing values for each column
print("Missing values per column:")
print(missing_values)
```

```python
# Get the count of male and female passengers
sex_counts = df['Sex'].value_counts()

# Display the counts
print("Count of passengers by sex:")
print(sex_counts)
```

```python
# Find the minimum and maximum values in the 'Fare' column
min_fare = df['Fare'].min()
max_fare = df['Fare'].max()

# Display the minimum and maximum fare
print(f"The minimum fare is: {min_fare:.2f}")
print(f"The maximum fare is: {max_fare:.2f}")
```

```python
# Calculate the average age of the passengers
average_age = df['Age'].mean()

# Display the average age
print(f"The average age of the passengers is: {average_age:.2f}")
```

```python
# Check the count of missing values in the 'Embarked' column
missing_embarked = df['Embarked'].isnull().sum()

# Display the count
print(f"Missing values in 'Embarked' column: {missing_embarked}")
```

```python
# Calculate the median age
median_age = df['Age'].median()

# Fill missing values in the 'Age' column with the median age using the recommended approach
df['Age'] = df['Age'].fillna(median_age)

# Verify that there are no more missing values in the 'Age' column
print("Missing values in 'Age' column after imputation:")
print(df['Age'].isnull().sum())
```

```python
# Plot a histogram of the 'Age' column
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
